<h2 class="alert alert-warning"><center>1. Tuples</center></h2>

Les tuples sont des séquences, assez semblables aux listes, sauf qu'on ne peut pas modifier un tuple après qu'il ait été créé. <br/>
Cela signifie qu'on définit le contenu d'un tuple (les objets qu'il doit contenir) lors de sa création, mais qu'on ne peut en ajouter ou en retirer par la suite.

<h3 class="alert alert-info"> Cours</h3>

### Définition d'un Tuple

```python
t = (1,2,3) # Triplet (Tuple à 3 éléments)
print(t, type(t))
```

```python
t = (1,) # Singleton (Tuple à un élément)
print(t, type(t))
```

### Accès aux éléments

```python
print(t[0], t[2])
```

### Modification d'un Tuple
On ne peut pas modifier directement un tuple 

```python
try :
    t[0] = 4
except :
    print("Impossible de modifier directement l'élément")

try : 
    t.append(4)
except : 
    print("Impossible de modifier le nombre d'éléments")
```

### Solution 
Il faut donc réaffecter le tuple

```python
t = (4,t[1],t[2])
print(t)
```

### Taille d'un tuple

```python
taille = len(t)
print(taille)
```

### Affectation multiple

```python
#Exemple 1
a,b = 3,5
print(a,b)
```

```python
#Exemple 2
a = 3
b = 5
b , a = a , b
print(a,b)
```

```python
def div_euclidienne(divid,divis):
    return divid//divis, divid%divis

q,r = div_euclidienne(18,7)
print(q,r)
resultat = div_euclidienne(18,7)
print(resultat, type(resultat))
```

<h2 class="alert alert-warning"><center>2. Dictionnaires</center></h2>

Les dictionnaires sont des séquences non ordonnées. Ils diffèrent des listes car les éléments sont indéxés non pas par des indices mais par des clés. 
Ces structures sont très utilisées car pratiques dans beaucoup de cas et très efficaces.

<h3 class="alert alert-info"> Cours</h3>

### Définition d'un Dictionnaire

```python
dico = {}
print(dico)
```

```python
dico ={'A': 1, 'B': 2}
print(dico, type(dico))
```

Dans le dictionnaire précédent `'A'` est une clé, `1` est la valeur associée à cette clé.

### Accès à une valeur

```python
dico['A']
```

```python
dico['C'] #Génère une erreur, la clé n'existe pas
```

### Modification et création de valeurs
Attention les instructions de modification et de création de valeurs sont les mêmes.<br/>
Si la clé existe on modifie la valeur, si elle n'existe pas, on la crée.

```python
dico['A']=0 # modification
dico['B']=1 # modification
dico['C']=2 # création
print(dico)
```

### Parcours de dictionnaires
voici deux méthodes utiles pour les dictionnaires : <br/>
`keys()` qui renvoie une structure itérable contenant toutes les clés du dictionnaire.<br/>
`items` qui renvoie une structure itérable contenant tous les tuples `(cle,valeur)` du dictionnaire.

```python
dico.keys()
```

```python
for k in dico.keys():
    print(k, dico[k])
```

```python
dico.items()
```

```python
for t in dico.items():
    print(t[0], t[1])
```

```python
for k,v in dico.items():
    print(k,v)
```
